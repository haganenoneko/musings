## Load data

```python
from seaborn import load_dataset
import pandas as pd 
iris = load_dataset('iris')
```

```python
iris.describe(include='all')
```

## Pivoting and other things

```python
iris.value_counts().sort_index(level=0)
```

```python
iris.value_counts().sort_index(level='sepal_length')
```

```python
iris.melt(id_vars=['sepal_length', 'sepal_width'], value_vars=['petal_length'])
```

```python
melted = iris.melt(id_vars=['sepal_length', 'sepal_width'], value_vars=['petal_length', 'petal_width'], var_name='petal_props')
melted 
```

```python
melted.pivot_table(index=['sepal_length', 'sepal_width'], columns='petal_props').\
    sort_index(level=0).reset_index()
```

```python
melted.pivot_table(index=['sepal_length', 'sepal_width'], columns='petal_props', aggfunc=sum).\
    sort_index(level=0).reset_index()
```

We can't use `pivot` because it doesn't aggregate values with duplicate indices.

```python
melted.pivot(index=['sepal_length', 'sepal_width'], columns='petal_props', values='value')
```
